# IF ,  ELIF, ELSE 

```python
marks = 75

if marks >= 90:
    print("Grade : A")
elif marks >= 60:
    print("Grade : B")
else:
    print("Grade: C, Need improvement!")
```

# For Loop
- used to iterate over sequences(lists, str, ranges)
- can perform arithemtic operation on specified range 

```python
for i in range(10):
    print("Indentation", i)
```

```python
for i in range(1,10):
    print("Indentation", i)
```

# While Loop
- Runs until the condition is false 

```python
count = 0

while count < 5:
    print("Count", count)
    count += 1
```

# simple Calculator 

```python
def calculator(a,b,op):
    if op == '+':
        print(a+b)
    elif op == '-':
        print(a-b)
    elif op == '*':
        print(a*b)
    elif op == '/':
        print(a/b)
    elif op == '//':
        print(a//b)
    elif op == '%':
        print(a%b)
    else:
        print('Invalid Operator!')
```

```python
calculator (2,4,'+')
```

```python
calculator (7,100,'*')
```
