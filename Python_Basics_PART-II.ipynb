# LISTS 
- A List is an ORDERED, MUTABLE(changable), collection of items.
- List can hold elements of different data types
# KEY FEATURES :
1. ORDERED( items have a fixed order)
2. MUTABLE(can change after creating the list)
3. Allows DUPLICATES
4. [ ] ~ this brackets are used 

```python
# CREATE A LIST 
fruits = ['Apple','Strawberry', 'Papaya']
print(fruits)

# SLICING 
print(fruits[0]) # first item of list
print(fruits[1]) # second
print(fruits[2]) # third
print(fruits[-1]) # last item we use '-'
```

```python
n =[1,2,3]
for i in n :
    print(i,i*2)
```

```python
fruits.append('mango') # adds the item at the last 
```

# TUPLES ( )

- A TUPLE is ORDERED, IMMUTABLE collection of items.
- Once created , the elements cannot be changed

## KEY FEATURES:
1. ORDERED
2. IMMUTABLE
3. Allows DUPLICATES
4. More memory efficient than lists

```python
# CREATE A TUPLE 
tup = (1,2,3,4,5)
```

```python
print(tup)
print(type(tup))
print(tup[3])
```

```python
tup.append(7) # You cannot change or add or delete 
```

# SETS { }
A SET is UNORDERED collection of unique items. 
It doesnot allow duplicates

# KEY FEATURES:
1. UNORDERED
2. NO DUPLICATES
3. MUTABLE

```python
# CREATE A SET
set_ = { 1,2,2,3,4,4}
print(set_) # IT REMOVES DUPLICATE VALUES N PRINTS UNIQUE VALUES
```

```python
set_.add(5)
```

```python
set_
```

```python
# SET OPERATIONS 
A = {1,6,3}
B = { 4,6,8}

print(A.union(B)) # unions both
print(A.intersection(B)) # common values in both sets 
print(A.difference(B)) # elements of 1 n not in 2 6 is in both sets so not printed 
```

# DICTIONARY {'KEY' : 'VALUE'}
A DICTIONARY  is a collection of key value pairs. 
Each key must be unique and is used to access its associated value

# KEY FEATURE
1. ORDERED( FROM PYTHON VERSION 3.7)
2. KEYS MUST BE UNIQUE AND IMMUTABLE
3. VALUES can be any type

```python
dic = {'Bob':23,'Alice': 24, 'Alex': 29,'Sam':25}
```

```python
dic
```

```python
type(dic)
```
